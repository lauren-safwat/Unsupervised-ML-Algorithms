### Imports

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.impute import KNNImputer
from sklearn.preprocessing import PowerTransformer, MinMaxScaler, StandardScaler, RobustScaler

import warnings
warnings.filterwarnings("ignore")
```

### 1.3 Outliers Identification

```python
Q1 = raw_data.quantile(0.25)
Q3 = raw_data.quantile(0.75)
IQR = Q3 - Q1
LB = Q1 - 1.5 * IQR
UB = Q3 + 1.5 * IQR
nOutliers = raw_data[(raw_data < LB) | (raw_data > UB)].count()
nOutliers
```

```python
f, axs = plt.subplots(figsize=(20, 20))
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.4, hspace=0.4)

for i, column in enumerate(raw_data.columns, 1):
  plt.subplot(6, 3, i)
  sns.boxplot(raw_data[column])
```

All data columns except PURCHASES_FREQUENCY and PURCHASES_INSTALLMENTS_FREQUENCY contain outliers

### 1.4 Skewness Identification

```python
f, axs = plt.subplots(figsize=(20, 20))
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.4, hspace=0.4)

for i, column in enumerate(raw_data.columns, 1):
  plt.subplot(6, 3, i)
  sns.distplot(raw_data[column], color="b", label="Skewness : %.1f"%(raw_data[column].skew()))
  plt.legend(loc="best")
```

Features have a skewed distributions

# **2. Data Preprocessing**

```python
preprocessed_df = raw_data.copy()
```

## **2.1 Handling missing values**

```python
imputer = KNNImputer(n_neighbors=5)
preprocessed_df = pd.DataFrame(imputer.fit_transform(preprocessed_df), columns=preprocessed_df.columns, index=preprocessed_df.index)
```

```python
preprocessed_df.isna().sum()
```

## **2.2 Data Transformation**

### **2.2.1 Log Transformation**

- The log transformation technique is primarily used to convert a skewed distribution to a normal distribution/less-skewed distribution. In this transform, we take the log of the values in a column and use these values as the column instead.

- Since the log of negative numbers and numbers between 0 and 1 is undefined, np.log1p is used to add 1 to all values before applying log transformation.

```python
lt_df = np.log1p(preprocessed_df)
```

```python
f, axs = plt.subplots(figsize=(20, 20))
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.4, hspace=0.4)

for i, column in enumerate(lt_df.columns, 1):
  plt.subplot(6, 3, i)
  sns.boxplot(lt_df[column])
```

```python
f, axs = plt.subplots(figsize=(20, 20))
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.4, hspace=0.4)

for i, column in enumerate(lt_df.columns, 1):
  plt.subplot(6, 3, i)
  sns.distplot(lt_df[column], color="b", label="Skewness : %.1f"%(lt_df[column].skew()))
  plt.legend(loc="best")
```

After the log transformation, some of the features have become more normally distributed while others still contain outliers and are skewed.
Also, there is no predetermined range for scaled data

### **2.2.2 Power Transformation**

- The Power Transformer also changes the distribution of the variable, as in, it makes it more Gaussian(normal). It actually automates the decision making of the best power transformer to use by introducing a parameter called lambda.

- It decides on a generalized power transform by finding the best value of lambda using either box-cox transform (works with strictly positive values only) or Yeo-Johnson transform (works with both positive and negative values).

- By default, zero-mean, unit-variance normalization is applied to the transformed data.

```python
pt = PowerTransformer()
pt_df = pd.DataFrame(pt.fit_transform(preprocessed_df), columns=preprocessed_df.columns, index=preprocessed_df.index)
```

```python
f, axs = plt.subplots(figsize=(20, 20))
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.4, hspace=0.4)

for i, column in enumerate(pt_df.columns, 1):
  plt.subplot(6, 3, i)
  sns.boxplot(pt_df[column])
```

```python
f, axs = plt.subplots(figsize=(20, 20))
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.4, hspace=0.4)

for i, column in enumerate(pt_df.columns, 1):
  plt.subplot(6, 3, i)
  sns.distplot(pt_df[column], color="b", label="Skewness : %.1f"%(pt_df[column].skew()))
  plt.legend(loc="best")
```

Here, the power transformer made most of the features more Gaussian-like and handled most of the outliers. So, it is the most suitable one in our case.

```python
pt_df.to_csv('Dataset/preprocessed_data.csv')
```

### **2.2.3 MinMax scaler**

- Min-Max Scaler scales all the data between 0 and 1 using the following formula:
x_scaled = (x  x_min)/(x_max  x_min)

```python
ms = MinMaxScaler()
ms_df = pd.DataFrame(ms.fit_transform(preprocessed_df), columns=preprocessed_df.columns, index=preprocessed_df.index)
```

```python
f, axs = plt.subplots(figsize=(20, 20))
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.4, hspace=0.4)

for i, column in enumerate(ms_df.columns, 1):
  plt.subplot(6, 3, i)
  sns.boxplot(ms_df[column])
```

```python
f, axs = plt.subplots(figsize=(20, 20))
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.4, hspace=0.4)

for i, column in enumerate(ms_df.columns, 1):
  plt.subplot(6, 3, i)
  sns.distplot(ms_df[column], color="b", label="Skewness : %.1f"%(ms_df[column].skew()))
  plt.legend(loc="best")
```

Here we can see that Min-Max scaler is sensitive to outliers and it doesnt reduce the skewness of a distribution.

### **2.2.4 Standard Scaler**

- Standardize features by removing the mean and scaling to unit variance.

```python
ss = StandardScaler()
ss_df = pd.DataFrame(ss.fit_transform(preprocessed_df), columns=preprocessed_df.columns, index=preprocessed_df.index)
```

```python
f, axs = plt.subplots(figsize=(20, 20))
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.4, hspace=0.4)

for i, column in enumerate(ss_df.columns, 1):
  plt.subplot(6, 3, i)
  sns.boxplot(ss_df[column])
```

```python
f, axs = plt.subplots(figsize=(20, 20))
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.4, hspace=0.4)

for i, column in enumerate(ss_df.columns, 1):
  plt.subplot(6, 3, i)
  sns.distplot(ss_df[column], color="b", label="Skewness : %.1f"%(ss_df[column].skew()))
  plt.legend(loc="best")
```

We can see that using the Standard scaler on these distributions shifted the mean to 0, but didnt really change the skewness of the data. This is because the outliers in this dataset have an impact on the feature mean and standard deviation used to scale the features.
